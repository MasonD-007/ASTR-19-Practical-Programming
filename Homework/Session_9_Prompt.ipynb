Use numpy to pull 1000 random numbers distributed uniformly between [0,1]. Histogram the random numbers into 100 bins, and plot the histogram. Label your axes and save the figure as a PDF.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.normal(scale=1.0, size=1000)
plt.hist(x, bins=100)
plt.xlabel('x')
plt.ylabel('count')
fig1 = plt.gcf()
plt.show()
fig1.savefig('histogram.pdf')
```
