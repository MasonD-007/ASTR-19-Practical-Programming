```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,2*np.pi,100)
y = np.sin(5*x)
```

```python
#changes the background color
plt.figure(facecolor="red")
#Plots the x and Y
plt.plot(x,y,)
#sets the y limit 
plt.ylim(-1.1,1.1)
#labels the x and y axis
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

fig1 = plt.gcf()
#plt.show()
fig1.savefig("demo.png",)
```
