```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,2*np.pi,100)
y = np.sin(5*x)
```

```python
fig = plt.figure()
plt.plot(x,y)
plt.ylim(-1.1,1.1)
```

```python
fig.gca().set_ylabel("Y (sin of X)")
fig.gca().set_xlabel("X")
fig
```

```python

fig.savefig("demo.png",facecolor='pink',)
```
