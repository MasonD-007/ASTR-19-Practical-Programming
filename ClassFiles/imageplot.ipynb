```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
```

```python
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111)
x = np.linspace(0, 1 * np.pi, 200)
y = np.linspace(0, 1 * np.pi, 200)
x, y = np.meshgrid(x, y)
z = -np.sin(x) * np.sin(y)
ax = plt.imshow(z, cmap=cm.turbo,origin='lower')
char = fig.colorbar(ax)
char.ax.get_yaxis().labelpad = 15
char.ax.set_ylabel('Z Value', rotation=270)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
