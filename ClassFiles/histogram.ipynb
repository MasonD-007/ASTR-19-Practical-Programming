```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 500

x = np.random.randn(n)

width = 0.5
histmin = np.floor(x.min())
histmax = np.ceil(x.max()) + width

bins = np.arange(histmin, histmax, width)
plt.hist(x, bins=bins,alpha=0.5, edgecolor='black')
plt.ylabel('N per bin')
plt.xlabel('x')
#plt.savefig('histogram.pdf', bbox_inches='tight',facecolor='white', transparent=False)
plt.show()
```
